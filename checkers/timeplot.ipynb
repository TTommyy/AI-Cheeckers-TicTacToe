```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
csv_file = "timing_results.csv"
orderd_data = pd.read_csv(csv_file, header=None, names=['Depth', 'Time'])
orderd_data
```

```python
csv_file = "random_timing_results.csv"
random_data = pd.read_csv(csv_file, header=None, names=['Depth', 'Time'])
random_data
```

```python
csv_file = "reverse_timing_results.csv"
reverse_data = pd.read_csv(csv_file, header=None, names=['Depth', 'Time'])
reverse_data
```

```python
csv_file = "min_max_timing_results.csv"
min_max_data = pd.read_csv(csv_file, header=None, names=['Depth', 'Time'])
min_max_data
```

```python

plt.figure(figsize=(6, 4))
plt.plot(orderd_data['Depth'], orderd_data['Time'], marker='o', label="orderd_data")
plt.plot(random_data['Depth'], random_data['Time'], marker='o', label="random_data")
plt.plot(reverse_data['Depth'], reverse_data['Time'], marker='o', label="reverse_data")
plt.plot(min_max_data['Depth'], min_max_data['Time'], marker='o', label="min_max_data")


plt.title('Alpha-Beta Search Performance')
plt.xlabel('Depth')
plt.ylabel('Time (seconds)')
plt.legend();
plt.grid(True)


plt.show()
```
