```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
csv_file = "timing_results.csv"
data = pd.read_csv(csv_file, header=None, names=['Depth', 'Time'])
data
```

```python

plt.figure(figsize=(6, 4))
plt.plot(data['Depth'], data['Time'], marker='o')

plt.title('Alpha-Beta Search Performance')
plt.xlabel('Depth')
plt.ylabel('Time (seconds)')
plt.grid(True)


plt.show()
```
